```python
import pandas as pd
import seaborn as sns
```

```python
file = pd.read_csv("/home/sai/Documents/GitHub/Python_Diwali_Sales_Analysis-main/Diwali Sales Data.csv",encoding="unicode_escape")
df = pd.DataFrame(file)
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
pd.isnull(df).sum()
```

```python
sns.countplot(x="Gender",data=df)
```

```python
ax = sns.countplot(x="Gender",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

Based on the analysis of the graph, it can be concluded that the frequency of orders placed by females surpasses that of males.

```python
df.columns
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(['Amount'],ascending=False)
```

```python
sales_amount = df.groupby(['Gender'],as_index=False)['Amount'].sum()
sns.barplot(x="Gender",y="Amount",data=sales_amount)
```

Based on the graph, we can conclude that females are spending a greater amount on orders compared to males.

```python
sns.countplot(x="Age Group",data=df,hue='Gender')
```

Based on the analysis of the graph, individuals in the 26-35 age group are placing more orders compared to other age groups, while those in the 55+ age group are ordering the least. Additionally, across all age groups, females are consistently placing more orders than males.

```python
age_sales = sns.countplot(x="Age Group",data=df,hue="Gender")
for bars in age_sales.containers:
    age_sales.bar_label(bars)
```

```python
df.groupby(['Age Group'],as_index=False)['Amount'].sum()
```

```python
age_group_sales = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=age_group_sales)
```

Based on the analysis of the graph, individuals in the 26-35 age group are spending the most, while those in the 0-17 age group are spending the least across all age groups.

```python
state_sales = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False).head()

sns.set(rc={'figure.figsize':(15,5)})

sns.barplot(x='State',y='Orders',data=state_sales)
```

The graph indicates that the majority of orders are coming from Uttar Pradesh, Maharashtra, Karnataka, Delhi, and Madhya Pradesh.

```python
state_orders=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False).tail()
sns.barplot(x="State",y='Orders',data=state_orders)
```

The graph also shows that the fewest orders are coming from Punjab and Telangana.

```python
maried_sales = sns.countplot(x='Marital_Status',data=df)
for bars in maried_sales.containers:
    maried_sales.bar_label(bars)
```

The graph indicates that unmarried individuals are placing more orders than married individuals.


```python
occupation_sales = sns.countplot(x='Occupation',data=df)
for ax in occupation_sales.containers:
    occupation_sales.bar_label(ax)
```

The graph indicates that individuals working in the IT sector, healthcare, and aviation are placing the most orders.

```python
occupation_sales = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values('Amount',ascending = False)
sns.barplot(x='Occupation',y='Amount',data=occupation_sales)
```

The graph indicates that individuals working in the IT sector, healthcare, and aviation are spending the most.

```python
occupation_sales = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values('Amount',ascending = False).head()
sns.barplot(x='Occupation',y='Amount',data=occupation_sales)
```

```python
product_sales = sns.countplot(x='Product_Category',data=df)
for bars in product_sales.containers:
    product_sales.bar_label(bars)
```

```python
product_sales = df.groupby(['Product_Category'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False).head()
sns.barplot(x='Product_Category',y='Orders',data=product_sales)
```

The graph indicates that most of the orders are in the Clothing & Apparel, Food, and Electronics & Gadgets categories.

```python
low_products = df.groupby(['Product_Category'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False).tail()
sns.barplot(x='Product_Category',y='Orders',data=low_products)
```

The graph indicates that the fewest orders are in the Veterinary, Tupperware, and Hand & Power Tools categories.

```python
top_products = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False).head()
sns.barplot(x='Product_Category',y='Amount',data=top_products)
```

The graph indicates that most of the spending is in the Food, Clothing & Apparel,and Electronics & Gadgets categories.

```python
low_products = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False).tail()
sns.barplot(x='Product_Category',y='Amount',data=low_products)
```

The graph indicates that the least spending is on Hand & Power Tools, Veterinary, and Office categories.
